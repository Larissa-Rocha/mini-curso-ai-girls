# Vamos ver como utilizar o python para realizar uma predição?

Vamos pensar juntos em uma tarefa de classificação e procurar por um algoritmo para fazermos o primeiro teste?

Um exemplo muito simples e muito conhecido para quem está começando no mundo dos dados é o Iris Dataset.

```python
# linear algebra
import numpy as np 

# data processing
import pandas as pd 

# data visualization
import seaborn as sns
%matplotlib inline
from matplotlib import pyplot as plt
from matplotlib import style

# Algorithms
from sklearn.datasets import load_iris
from sklearn import tree
from sklearn.metrics import accuracy_score


```

```python

X, y = load_iris(return_X_y=True)
```

```python
iris = load_iris()
data1 = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                     columns= iris['feature_names'] + ['target'])
```

```python
data1.head()
```

```python
data1.shape
```

```python
data1.target.unique()
```

<img src='figures/iris-machinelearning.png'>

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X, y)
```

```python
tree.plot_tree(clf) ;
```

```python
# iris.target_names
```

```python
fn=['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']
cn=['setosa', 'versicolor', 'virginica']
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (4,4), dpi=300)
tree.plot_tree(clf,
               feature_names = fn, 
               class_names=cn,
               filled = True);
fig.savefig('imagename.png')
```

```python
y_pred = clf.predict(X)
```

```python
y_pred
```

```python
accuracy_score(y, y_pred, normalize=True)
```

o que estamos vendo aqui: 
Esse algoritmo busca pelas variáveis mais importantes e de acordo com a combinação delas, nos diz a qual classe um objeto pertence;
